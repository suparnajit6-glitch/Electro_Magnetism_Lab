```python
# Here I am going to do some sandbox stuffs on electric field and its relationship with potential

#Some basic plots on Gradient of a scaler fn()




import sympy as sp 
from sympy.vector import CoordSys3D,gradient
import numpy as np
import matplotlib.pyplot as plt




N=CoordSys3D('N')

x,y,z=N.base_scalars()

#position of slits

s1=(0,5)
s2=(0,-5)


r1 = sp.sqrt((x-s1[0])**2 + (y-s1[1])**2) 
o1= sp.sin(r1)**2

r2 = sp.sqrt((x-s2[0])**2 + (y-s2[1])**2)
o2= sp.sin(r2)**2



v1=gradient(o1,N)

vi1=v1.dot(N.i)
vj1=v1.dot(N.j)
vk1=v1.dot(N.k)

v2=gradient(o2,N)

vi2=v2.dot(N.i)
vj2=v2.dot(N.j)
vk2=v2.dot(N.k)

fx1=sp.lambdify((x,y,z),vi1,'numpy')
fy1=sp.lambdify((x,y,z),vj1,'numpy')
fz1=sp.lambdify((x,y,z),vk1,'numpy')

fx2=sp.lambdify((x,y,z),vi2,'numpy')
fy2=sp.lambdify((x,y,z),vj2,'numpy')
fz2=sp.lambdify((x,y,z),vk2,'numpy')


def f(x,y,z):
    a1 = np.sqrt(fx1(x,y,z)**2 + fy1(x,y,z)**2 + fz1(x,y,z)**2)
    a2 = np.sqrt(fx2(x,y,z)**2 + fy2(x,y,z)**2 + fz2(x,y,z)**2)
    return a1+a2


y=np.linspace(-20,20,1000)
x=np.linspace(0,40,1000)

X,Y=np.meshgrid(x,y)

Z=0




plt.contourf(X,Y,f(X,Y,Z),levels=500,cmap="plasma")

plt.grid(True,color="black")
plt.colorbar()

```

```python
plt.plot(y,f(40,y,0),color="brown")
plt.grid(True)
```

```python
# Here I am going to do some sandbox stuffs on electric field and its relationship with potential

#Some basic plots on Gradient of a scaler fn()




import sympy as sp 
from sympy.vector import CoordSys3D,gradient
import numpy as np
import matplotlib.pyplot as plt




N=CoordSys3D('N')

x,y,z=N.base_scalars()

#position of slits

s1=(0,5,0)
s2=(0,-5,0)

ω=2*sp.pi


r1 = sp.sqrt((x-s1[0])**2 + (y-s1[1])**2 + (z-s1[2])**2) 
o1= sp.sin(ω*r1)

r2 = sp.sqrt((x-s2[0])**2 + (y-s2[1])**2 + (z-s2[2])**2)
o2= sp.sin(ω*r2)

v1=sp.lambdify((x,y,z),o1,"numpy")
v2=sp.lambdify((x,y,z),o2,"numpy")






def f(x,y,z):

    a=v1(x,y,z)+v2(x,y,z)

    return a


y=np.linspace(-10,10,1000)
x=np.linspace(0,100,1000)

X,Y=np.meshgrid(x,y)

Z=0




plt.contourf(X,Y,f(X,Y,Z),levels=500,cmap="magma")

plt.plot(s1[0],s1[1])
plt.plot(s2[0],s2[1])

plt.grid(True,color="black")
plt.colorbar()

```

```python

plt.style.use('dark_background')
plt.plot(y,f(100,y,Z)**2,color="red")


plt.grid(True,color="white",alpha=0.4)
```
