```python
# Here I am going to do some sandbox stuffs on electric field and its relationship with potential

#Some basic plots on Gradient of a scaler fn()




import sympy as sp 
from sympy.vector import CoordSys3D,gradient
import numpy as np
import matplotlib.pyplot as plt




N=CoordSys3D('N')

x,y,z=N.base_scalars()

r = sp.sqrt(x**2 + y**2)
o = sp.sin(r)


v=gradient(o,N)

vi=v.dot(N.i)
vj=v.dot(N.j)
vk=v.dot(N.k)

fx=sp.lambdify((x,y,z),vi,'numpy')
fy=sp.lambdify((x,y,z),vj,'numpy')
fz=sp.lambdify((x,y,z),vk,'numpy')


def f(x,y,z):
    a = np.sqrt(fx(x,y,z)**2 + fy(x,y,z)**2 + fz(x,y,z)**2)

    return a 


x=y=np.linspace(-20,20,1000)

X,Y=np.meshgrid(x,y)

Z=0




plt.contourf(X,Y,f(X,Y,Z),levels=500,cmap="plasma")
plt.colorbar()

```
